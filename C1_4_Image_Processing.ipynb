Image processing and compression using convolution

```
import cv2
import numpy as np
from scipy import misc
i=misc.ascent()
```

```
import matplotlib.pyplot as plt
plt.grid(False)
plt.gray()
plt.axis("off")
plt.imshow(i)
plt.show()
```

```
i_transformed=np.copy(i)
size_x=i_transformed.shape[0]
size_y=i_transformed.shape[1]
```

```
#filter=np.array([[0,1,0],[1,-4,1],[0,1,0]])
filter = np.array([ [-1, -2, -1], [0, 0, 0], [1, 2, 1]])
#filter = [ [-1, 0, 1], [-2, 0, 2], [-1, 0, 1]]
```

```
for x in range(0,size_x-2):
  for y in range(0,size_x-2):
    convolution=np.sum(i_transformed[x:x+3,y:y+3]*filter)
    if convolution<0:
      convolution=0
    if (convolution>255):
      convolution=255.0
    i_transformed[x,y]=convolution
```

```
plt.gray()
plt.grid(False)
plt.imshow(i_transformed)
plt.show()
```

```
new_x=int(size_x/2)
new_y=int(size_y/2)
newimage=np.zeros((new_x,new_y))
for x in range(0,size_x,2):
  for y in range(0,size_y,2):
    newimage[int(x/2),int(y/2)]=np.max(i_transformed[x:x+2,y:y+2])
```

```
plt.gray()
plt.grid(False)
plt.imshow(newimage)
plt.show()
```
